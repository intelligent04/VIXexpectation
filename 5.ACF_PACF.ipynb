```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

residual_charts = ['./residual_data/DIA_residual.csv','./residual_data/Gold_residual.csv','./residual_data/Oil_residual.csv',
'./residual_data/QQQ_residual.csv','./residual_data/SPY_residual.csv',
'./residual_data/US10-Year_Bond_residual.csv','./residual_data/USD_KRW_residual.csv','./residual_data/VIX_residual.csv']

normalized_charts =['./normalized_data/DIA_residual_normalized.csv','./normalized_data/Gold_residual_normalized.csv','./normalized_data/Oil_residual_normalized.csv',
'./normalized_data/QQQ_residual_normalized.csv','./normalized_data/SPY_residual_normalized.csv',
'./normalized_data/US10-Year_Bond_residual_normalized.csv','./normalized_data/USD_KRW_residual_normalized.csv','./normalized_data/VIX_residual_normalized.csv']

indices=['DIA','Gold','Oil','QQQ','SPY','US10-Year_Bond','USD_KRW','VIX']


for i in range(8):
    df1 = pd.read_csv(residual_charts[i], parse_dates=['Date'], index_col='Date')  # 정상화 전 차트
    df2 = pd.read_csv(normalized_charts[i], parse_dates=['Date'], index_col='Date')  # 정상화 후 차트
    
    # 2. ACF/PACF 비교 시각화
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))
    
    # 정상화 전 ACF
    plot_acf(df1['resid'].dropna(), lags=50, ax=ax1, alpha=0.05, title='')
    ax1.set_title(f'{indices[i]} - original residual (ACF)', fontsize=14)
    
    # 정상화 전 PACF
    plot_pacf(df1['resid'].dropna(), lags=50, ax=ax2, alpha=0.05, title='')
    ax2.set_title(f'{indices[i]} - original residual (PACF)', fontsize=14)
    
    # 정상화 후 ACF
    plot_acf(df2['diff_log_resid'].dropna(), lags=50, ax=ax3, alpha=0.05, title='',color="red")
    ax3.set_title(f'{indices[i]} - after normalization (ACF)', fontsize=14)
    
    # 정상화 후 PACF
    plot_pacf(df2['diff_log_resid'].dropna(), lags=50, ax=ax4, alpha=0.05, title='',color="red")
    ax4.set_title(f'{indices[i]} - after normalization (PACF)', fontsize=14)
    
    plt.tight_layout()
    plt.show()
    
    
    plt.tight_layout()
    plt.show()
```
