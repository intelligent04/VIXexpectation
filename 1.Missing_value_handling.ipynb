## (2) DIA

```python
# 데이터셋 로드
file_path = "./data/DIA ETF Stock Price History.csv"
df_dia = pd.read_csv(file_path)

# 기본 정보 표시 그리고 결측치 확인 
df_dia.info(), df_dia.isnull().sum()
```

```python
missing_rows_dia = df[df.isnull().any(axis=1)]
missing_rows_dia
```

공휴일인가?

```python
import holidays

# 미국 공휴일 리스트 가져오기
us_holidays = holidays.US()

# 결측치가 포함된 행을 깊은 복사본으로 생성
missing_rows_dia = df_dia[df_dia.isnull().any(axis=1)].copy()

# Convert the 'Date' column to datetime format
missing_rows_dia.loc[:, "Date"] = pd.to_datetime(missing_rows_dia["Date"], format="%m/%d/%Y")

# 함수: 날짜가 미국 공휴일인지 확인
def is_us_holiday_or_weekend(date):
    return date in us_holidays or date.weekday() >= 5  # 주말(토요일: 5, 일요일: 6) 포함

# 공휴일 또는 주말 여부 판단
missing_rows_dia.loc[:, "Is_Holiday_or_Weekend"] = missing_rows_dia["Date"].apply(is_us_holiday_or_weekend)

# 결과 확인
missing_rows_dia
```

```python
df_dia_cleaned = df_dia.dropna()

cleaned_file_path = "./processed_data/DIA_ETF_Stock_Price_History.csv"
df_dia_cleaned.to_csv(cleaned_file_path, index=False)
df_dia_cleaned.info(), df_dia_cleaned.isnull().sum()
```

## (5) Gold

```python
# 데이터셋 로드
file_path = "./data/Gold Futures Historical Data.csv"
df_gold = pd.read_csv(file_path)

# 기본 정보 표시 그리고 결측치 확인 
df_gold.info(), df_gold.isnull().sum()
```

```python
# 결측치가 포함된 행을 깊은 복사본으로 생성
missing_rows_gold = df_gold[df_gold.isnull().any(axis=1)].copy()

# Convert the 'Date' column to datetime format
missing_rows_gold.loc[:, "Date"] = pd.to_datetime(missing_rows_gold["Date"], format="%m/%d/%Y")

# 함수: 날짜가 미국 공휴일인지 확인
def is_us_holiday_or_weekend(date):
    return date in us_holidays or date.weekday() >= 5  # 주말(토요일: 5, 일요일: 6) 포함

# 공휴일 또는 주말 여부 판단
missing_rows_gold.loc[:, "Is_Holiday_or_Weekend"] = missing_rows_gold["Date"].apply(is_us_holiday_or_weekend)

# 결과 확인
missing_rows_gold
```

- 2013년 8월 13일은 금요일임
- Bloomberg에서 찾아봐도 보이지 않음
- 선형 보간법 사용하자

1. `df_gold`를 `datatimeindex`로 바꾼다다
2. 거래량(Vol) 칼럼을 object에서 float으로 바꿔야 한다

```python
# Ensure 'Date' is set as index in datetime format
df_gold["Date"] = pd.to_datetime(df_gold["Date"], format="%m/%d/%Y")
df_gold.set_index("Date", inplace=True)

# Convert 'Vol.' column to numeric format
def convert_volume(vol):
    if isinstance(vol, str):
        vol = vol.replace("K", "e3").replace("M", "e6")  # Convert K/M to numeric format
        try:
            return float(eval(vol))  # Convert expressions like "105.72e3" -> 105720
        except:
            return None  # Return None for non-convertible values
    return vol

df_gold["Vol."] = df_gold["Vol."].apply(convert_volume)

# Apply time-weighted interpolation method now that 'Vol.' is numeric and index is datetime
df_gold["Vol."] = df_gold["Vol."].interpolate(method="time")

# Reset index after interpolation
df_gold.reset_index(inplace=True)

# Save the corrected dataset
cleaned_file_path_corrected = "./processed_data/Gold_Futures_Interpolated.csv"
df_gold.to_csv(cleaned_file_path_corrected, index=False)
```
